### 2. 재현 인물검색 html 가지고오기

```python
url = "https://people.search.naver.com/search.naver?where=nexearch&query=%EC%9E%AC%ED%98%84&sm=tab_etc&ie=utf8&key=PeopleService&os=1691787"

headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_5) AppleWebKit 537.36 (KHTML, like Gecko) Chrome",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8"
}
```

```python
url
```

```python
response = requests.get(url, headers=headers)
source = response.text
source
```

```python
soup = BeautifulSoup(source, 'lxml')
soup
```

```python
soup.find('meta')
```

```python
soup.find_all('meta')
```

```python
soup.prettify
```

```python
# tit_main _text
# 앨범 이름 뽑기
# 공연 이름뽑기
# 한 화면에 5개씩
soup.find_all('span', attrs={'class' : 'tit_main _text'})
```

```python
# status_sub
# 앨범의 가수 이름 & 날짜 뽑기
# 공연의 가수 이름 & 위치 뽑기
# 한 화면에 5개씩
soup.find_all('span', attrs={'class' : 'status_sub'})
```

```python
soup.find_all('span', attrs={'class' : 'status_sub _text'})
```

```python
soup.find_all('span', attrs={'class' : 'status_sub'})[6].text
```

```python
# 앨범명, 가수명, 날짜
# 공연명, 위치, 
tit_list = []

for i in range(0,10):
    title = soup.find_all('span', attrs={'class' : 'tit_main _text'})[i].text
    sub_title = soup.find_all('span', attrs={'class' : 'status_sub _text'})[i].text
    sub_title_date = soup.find_all('span', attrs={'class' : 'status_sub'})[i*2+1].text
    data = title, sub_title, sub_title_date
    tit_list.append(data)
```

```python
tit_list
```

### 4. csv로 내보내기

```python
import pandas as pd
```

```python
# 앨범명, 가수명, 날짜
tit_album = tit_list[0:5]
tit_album
```

```python
tit_album = pd.DataFrame(tit_album, columns =['album_title','singer','date'])
```

```python
tit_album.to_csv('재현 인물정보 앨범.csv',encoding ='utf-8')
```

```python
# 공연명, 위치, 
tit_perform = tit_list[5:10]
tit_perform
```

```python
tit_perform = pd.DataFrame(tit_perform, columns =['perform_title','where',''])
```

```python
tit_perform.to_csv('재현 인물정보 공연.csv',encoding ='utf-8')
```
