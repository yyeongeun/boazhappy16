```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("C:/Users/김영은/Desktop/iris.csv")
```

```python
dataset
```

```python
#scatter plot
import matplotlib.pyplot as plt

plt.scatter(dataset.SepalLength,dataset.SepalWidth)
plt.title("Scatter plot")
plt.xlabel("SepalLength")
plt.ylabel("SepalWidth")
```

```python
#correlation coefficient
r=np.corrcoef(dataset.SepalLength,dataset.SepalWidth)
print(r)
#약한 음의 상관관계 확인 가능
```

```python
#모든 변수간 산점도 : pairplot()
import seaborn as sns
cols=['SepalLength','SepalWidth','PetalLength','PetalWidth']
dataset1=dataset[cols]
```

```python
sns.pairplot(dataset1)
```

```python
import statsmodels.formula.api as sm
model = sm.ols(formula='SepalLength~SepalWidth+PetalLength+PetalWidth', 
               data=dataset).fit()
print(model.summary())
#적합된 회귀식
#SepalLength = 1.8451 + 0.6549 SepalWidth + 0.7111 PetalLength - 0.5626 PetalWidth
```

```python
#임의의 값에 대하여 SepalLength의 예측값 구하기
from sklearn.linear_model import LinearRegression
model=LinearRegression()
x=dataset[['SepalWidth','PetalLength','PetalWidth']]
y=dataset[['SepalLength']]
model.fit(x,y)
SepalWidth=3.9
PetalLength=1.7
PetalWidth=0.4

X=[SepalWidth,PetalLength,PetalWidth]
prediction=model.predict([X])
print(prediction)
```
